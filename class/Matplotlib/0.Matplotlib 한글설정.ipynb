# 한글폰트 설치

```python
# 일단 import 
%matplotlib inline  

import matplotlib as mpl  # 기본 설정 만지는 용도
import matplotlib.pyplot as plt  # 그래프 그리는 용도
import matplotlib.font_manager as fm  # 폰트 관련 용도
```

## 한글 깨지는 것 확인하기

```python
# 데이터 준비
import numpy as np

data = np.random.randint(-200, 100, 50).cumsum()
data
# 한글을 넣어놓고 그러보면 깨진다
plt.figure(figsize=(6,4))
plt.plot(range(50), data, 'r')
plt.title('시간별 가격추이')
plt.ylabel('주식가격')
plt.xlabel('시간(분)')
```

## 시스템에 설치된 폰트 확인.

```python
sys_font=fm.findSystemFonts()
print(f"sys_font number: {len(sys_font)}")
print(sys_font)

font = [f for f in sys_font if 'Nanum' in f]  #나눔폰트찾기
print(f"nanum_font number: {len(font)}")

font
```

```python
font = [f for f in sys_font if 'malgun' in f]  #맑은폰트찾기
print(f"Malgun_font number: {len(font)}")
font
```

```python
# matplotlib 한글 폰트 오류 문제 해결 

import matplotlib.font_manager as fm
import matplotlib.pyplot as plt

#폰트파일위치 지정해서 설정
font_path = './NanumSquareR.otf'
# font_path = 'Users\\bag-yeju\\Desktop\\Dev_mulcam\\NanumSquare\\NanumSquareEB.otf'
font_name = fm.FontProperties(fname=font_path).get_name()

#폰트 설정
plt.rc('font', family=font_name)

# 마이너스 깨짐 현상 해결
plt.rcParams['axes.unicode_minus'] = False

# matplotlib 폰트 패밀리 출력
plt.rcParams['font.family']
```
