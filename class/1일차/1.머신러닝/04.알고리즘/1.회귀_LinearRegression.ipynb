## 회귀 개요

* 회귀는 현대 통계학을 이루는 큰 축입니다. 
* 그런데, 회귀라는 말 자체가 친근하지 않은 이상하게 들리는 말입니다. 도대체 뭘 회귀한다는 건지...


* **"부모의 키가 크더라도 자식의 키가 대를 이어 무한정 커지지 않으며, 부모의 키가 작더라도 대를 이어 자식의 키가 무한정 작아지지 않는다"**
* 회귀 분석은 이처럼 데이터 값이 **평균**과 같은 일정한 값으로 돌아가려는 경향을 이용한 통계학 기법입니다. 


* 회귀는 여러개의 피처(독립변수)와 한 개의 레이블(종속변수) 간의 상관관계를 모델링하는 기법입니다. 
* 연속형 값을 예측하기 위한 모델

## Linear Regression(선형회귀)

* 기본적인 머신러닝 기법으로 100년전 개발된 모델이며, 개선된 다양한 모델이 존재합니다. 
* 단순 선형 모델은 **```성능이 아쉽지만```**, 활용을 위해 가볍게 컨셉만 이해합시다.

## Polynomial Regression(다항 회귀)

* 선형회귀는 1차원식이고, 다항회귀는 2차,3차 방정식과 같은 다항식을 쓰는 모델입니다. 
* 회귀선이 곡선입니다. 

image.png

* 단항 피처[$x_1,x_2$]를 Degree=2, 즉 2차 다항 피처로 변환하면?
    * $(x_1 + x_2)^2$의 식 전개에 대응되는 [$1, x_1, x_2, x_1x_2, x_1^2, x_2^2$]의 다항 피처들로 변환


* 단항 피처[$x_1,x_2$]를 Degree=3, 즉 3차 다항 피처로 변환하면?
    *  $(x_1 + x_2)^3$의 식 전개에 대응되는 [$1, x_1, x_2, x_1x_2, x_1^2, x_2^2, x_1^3, x_1^2x_2, x_1x_2^2, x_2^3$]의 다항 피처들로 변환

## 모델 적합 과정 : 훈련을 한다는 것은 뭐지?

* 공부시간(입력)과 시험성적(출력)과의 관계를 잘 설명할 수 있는 선(모델)을 찾습니다. 
* 여러 직선 중 실제값과 모델 사이의 오류값의 합이 최소가 될 수 있는 최적의 회귀 계수를 찾습니다. (누가? 모델이 자동으로)
* 오류값의 합을 RSS(residual sum of squares)라하고 **```비용함수(cost function)```** 또는 **```손실함수(Loss function)```** 라고 합니다. 
* 손실함수 즉, 모델이 얼마나 엉터리인지에 대한 값입니다. 
* 손실함수가 점진적으로 작아지도록 수렴해 가는 과정을 **```경사하강법```** 이란 기법을 사용하여 모델에 구현되어 있다.

## 경사하강법(Gradient Descent)

* 점진적으로 반복적인 계산을 통해 W 파라미터 값을 업데이트하면서 오류값이 최소가 되는 W 파라미터값을 구하는 방식입니다. 

* 최초 오류값이 100었다면, 두번째 오류값은 100보다 작은 90, 세번째는 80으로 지속적으로 오류를 감소시키는 방향으로 W값 업데이트합니다. 
* 그럼 어떻게 오류를 보정할까?
* 편미분 결과값을 반복적으로 보정하면서 w1,w0값을 업데이트합니다.

## Overfit(과적합)

* 훈련 데이터에만 딱 맞아서, 일반적인 다른 데이터에서는 적절한 예측이 어려운 경우를 말합니다. 
* 해결방법
    * 특성 갯수 줄이기
        * 주요한 특징을 가진 feature만 사용하고, 나머지는 버리기
        * 필요시 주요한 특징을 가진 feature를 직접 선택하거나 추가하고, 중요도에 따라 나머지는 삭제
    * feature값에 대한 정규화(regularization) 수행

## 회귀에 사용되는 데이터 형태

* 회귀 모델과 같은 선형 모델은 **```피처와 레이블값 간에 선형관계```** 가 있다고 가정하고, 
* 이러한 최적의 선형함수를 찾아내 결과값을 예측합니다. 
* **```피처값과 레이블값의 분포가 정규 분포```** (평균을 중심으로 종 모양으로 데이터 값이 분포된 형태)형태를 매우 선호합니다. 
    * 레이블은 반드시 정규분포를 가져야 함, 아니면 로그변환 적용
    * 피처는
        * StandardScaler()클래스 사용: 평균이 0, 분산이 1인 표준 정규 분포를 가지도록 변환
        * MinMaxScaler()클래스 사용: 최소값 0, 최대값 1로 정규화 수행
        * 로그 변환


image.png

* 각 **```피처간에는 독립```** 을 가정합니다. 
* 피처간의 상관관계가 매우 높은 경우 분산이 매우 커져서 오류에 매우 민감해 집니다. 
* 이러한 현상을 다중공선성문제라고 합니다. 
* 대표 피처만 남기고 제거하거나 규제합니다.
