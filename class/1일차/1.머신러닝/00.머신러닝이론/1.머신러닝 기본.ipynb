# 기계학습 모델

* 문제도 주고 답도 주는 지도학습
* 지도학습에는 2가지 종류가 있습니다.
* 어떤것을 예측하는가에 따라  **```회귀(Regression)```** 와 **```분류(Classification)```** 가 있습니다. 
* 범주형 데이터를 예측할 경우: 분류
* 수치형 데이터를 예측할 경우: 회귀
* 비지도 학습 모델에는 **```군집(Clustering)```** 이 있습니다. 

## 지도학습:예측(회귀)

* 주어진 입력과 출력의 쌍을 학습한 후,
* 새로운 데이터를 입력했을 때,
* 합리적인 출력값을 예측하는 것입니다. 
* 훈련 데이터는 **```연속적인 값```** 으로 입력과 출력이 모두 **```실수```** 입니다. 
* 예)김장철 배추 가격 예측, 상품 판매량 예측, 날씨에 대한 예측, 주택 가격의 예측

## 지도학습:분류

* 훈련 데이터에 각각 레이블을 지정하여 학습하여 분류합니다. 
* 예측에서는 연속적인 값을 출력했지만,
* 분류에서는 연속되지 않고 **```단절되어 있는 값```** 을 출력합니다. 
* 무게,브릭스 농도를 보고 출하여부를 분류 예측하면 이항 분류
* 무게,브릭스 농도를 보고 상,중,하로 분류 예측하면 다항 분류
* 예) 스팸메일 분류, 손 글씨 분류, 암환자 분류, 0~9까지의 숫자 구분

```
[참고] 분류와 회귀 모델을 기계는 어떻게 표현할까?
```
**분류 모델**
* 원형과 십자형 범주형 데이터가 같은 평면상에 분포되어 있습니다. 
* 분류 모델은 2가지 형태의 데이터를 구분하는 경계를 나누는 선으로 표현됩니다. 
* 이 평면상의 선은 수학적으로는 함수로 표현될 수 있다는 것입니다. 
* 즉, 범주형 데이터를 구분하는 분류 모델을
기계는 어떠한 함수로 표현할 수 있습니다. 

**회귀 모델**
* 수치형 데이터가 평면상에 분포되어 있습니다. 
* 증가하고 감소하는 데이터의 규칙을 찾고 오차를 가장 적게 하는 선을 긋는 것으로 표현됩니다. 
* 즉, 회귀 모델은 데이터들을 가장 잘 대변하는 어떠한 함수로 표현될 수 있습니다. 

**주의: 예제에서는 이해하기 쉽게 직선으로 표현되었지만, 실제 세계에서 발생하는 데이터는 대부분 비선형(곡선)으로 표현됩니다.  




## 비지도 학습:군집

* 문제만 주는 비지도 학습
* 즉, 답(레이블)이 없이 학습하는 학습법
* 스스로가 데이터 간의 패턴을 찾아 분류 기준을 찾고, 비슷한 유형의 데이터 끼리 그룹핑하면서 군집을 만듭니다.  

image.png

[활용 예시]
* 직원 A
    * 자기 경험과 직관에 따라 20~30대 고객이 운동화를 많이 구매한다는 것을 인지
    * 이 고객군을 타깃 마케팅 방식을 제안
* 직원 B
    * 비슷한 경험이 있지만 그간의 판매 정보를 데이터화 하여 비지도 학습에 적용해 봄
    * 운동화 매출에 큰 영향을 준 고객군에는 마라톤, 등산과 같이 외부 활동을 즐기는 고객이거나
    * 스트리트 패션 커뮤니티를 많이 애용하는 고객인 사례를 식별할 수 있었음.
    * 조금더 면밀하고 명확한 타깃 마케팅이 가능

* 정답 데이터를 만들지 않아도
* 보유한 데이터 간의 관계나 유사성을 발굴할 수 있다는 점이 장점
* 정답 데이터가 없으므로 예측 결과를 검증할 수 없다는 점이 단점

# 데이터 속성과 유형

* 데이터 속성은 다른 것과 구별할 수 있는 성질을 말합니다. 
* 데이터는 관찰이나 실험을 통해 수집되는 값 또는 특성을 의미하며, 하나의 데이터는 여러가지 **속성** 들로 구성됩니다. 
* 붓꽃 데이터 속성(Feature)
    * Sepal_length(꽃받침 길이)
    * Sepal_width(꽃받침 너비)
    * Petal_length(꽃잎 길이)
    * Petal_width(꽃잎 너비)

* 붓꽃 품종은 (Class)
    * Setosa
    * Versicolor
    * Virgnica

    

* 데이터 자료형

    * 연속형 데이터 : 꽃잎의 길이와 너비는 실수 형태의 값을 가지는 **연속형 데이터** 입니다. 
    * 범주형 데이터 : 꽃의 품종은 문자 형태의 정해진 값을 가지는 **범주형 데이터** 입니다.  

* Feature와 Class의 관계를 이야기 할 때

    * 붓꽃의 꽃받침 길이와 너비, 꽃잎의 길이와 너비가 품종에 어떤 영향을 미치는지 관계를 파악할 때,
    * 결과에 영향을 미치는 원인이 되는 변수를 **```독립변수```** 라고 합니다. (Sepal_length, Sepal_width, Petal_length, Petal_width) 
    * 특성값에 따라 결정되는 변수를 **```종속변수```** 라고 합니다.  (품종)

## 데이터 속성

* 피처(Feature)
    * 피처는 데이터 세트의 일반 속성입니다. 
    * 레이블을 제외한 나머지 **```속성```** 을 모두 피처로 지칭합니다. 
    
* 레이블(Label)
    * 데이터의 학습을 위해 주어진 **```정답```** 데이터입니다.

## 언더 피팅 VS 오버 피팅

* 머신러닝의 최종 목표는 **일반화**에서도 좋은 성능을 내는 것입니다. 
    * 고3이 공부하는 이유는 모의고사 만점이 아니라, 
    * 어떤 문제가 출제될 지 모르는 수능에서의 만점이 목표입니다.

* **언더 피팅(Underfitting)**
    * 학습이 부족한 상태
    * 지나친 단순화로 인해 에러가 많이 발생한 상태

* **오버 피팅(Overfitting)**
    * 학습이 너무 많이 된 상태
    * 학습 데이터에 대해 과하게 학습되어, 학습 데이터 이외의 데이터에 대해서는 잘 동작하는 못하는 상태
