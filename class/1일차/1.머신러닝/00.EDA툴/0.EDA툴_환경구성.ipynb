## 파이썬 가상환경 만들기


1.현재 작업폴더 (00.EDA툴)생성

2.터미널 열고 (현재 위치 예:C:\Dev_mulcam1\1.머신러닝\00.EDA툴 )로 이동

3.아래의 코드로 파이썬 가상환경 만들기
```
python -m venv venv
```

4.가상환경 활성화

```
cd venv
Scripts\activate
```


5.pip 업그레이드
```
pip list
```



```
python.exe -m pip install --upgrade pip
```

5.sweetviz 설치
```
pip install sweetviz
``````

```
pip show sweetviz
```
