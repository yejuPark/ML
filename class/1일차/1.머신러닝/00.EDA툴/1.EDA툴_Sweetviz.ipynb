## sweetviz 사용해보기

### 단일 분석

```python
import pandas as pd
import sweetviz as sv
# 경고메세지 끄기
import warnings
warnings.filterwarnings(action='ignore')

df_train = pd.read_csv('./data/train.csv')

my_report = sv.analyze(df_train)
# my_report.show_notebook(layout='widescreen',scale=0.8) #notebook, colab에서 표시하기
my_report.show_html(r'./image/sweet_report.html')
```

### 비교 분석

분석 해석
https://towardsdatascience.com/powerful-eda-exploratory-data-analysis-in-just-two-lines-of-code-using-sweetviz-6c943d32f34

```python
import pandas as pd
import sweetviz as sv

# 경고메세지 끄기
import warnings
warnings.filterwarnings(action='ignore')

df_train = pd.read_csv('./data/train.csv')
df_test = pd.read_csv('./data/test.csv')

compare_report = sv.compare([df_train,'Train'],[df_test,'Test'],target_feat='Survived')
# compare_report.show_notebook(layout='widescreen',scale=0.8) #notebook, colab에서 표시하기
compare_report.show_html(r'./image/sweet_report.html')
```

## 한글 설정

```python
import pandas as pd
import sweetviz as sv
# 경고메세지 끄기
import warnings
warnings.filterwarnings(action='ignore')
# sv.config_parser.read("Override.ini")

sv.config_parser.read_string("[General]\nuse_cjk_font=1") #중국어,일본어,한국어 활성

df = pd.DataFrame({'1열': [1,2,3], '과일': ['사과', '배', '바나나']}, dtype='category')
my_report = sv.analyze(df)
my_report.show_html(r'./image/sweet_report_test.html')
```
