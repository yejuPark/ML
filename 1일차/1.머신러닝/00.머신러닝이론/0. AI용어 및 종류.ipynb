# 인공지능/머신러닝/딥러닝

* 인공지능(Artificail Intelligence): 사람처럼 똑똑한 기계
* 머신러닝(Machine Learning): 인공지능을 구현하는데 성공한 방법, 기계학습
* 딥러닝(Deep Learning): 머신러닝보다 더 똑똑한 인공지능을 구현하는 데 성공한 방법, 인간의 뉴런과 비슷한 인공신경망 기반

## 학습이란?

* 많은 자료를 보고 외우고 틀린 문제를 반복하면서 오답을 최소화하는 것 
* 이와 같이 머신러닝도 사람이 제공한 데이터에 대해서 문제를 반복해서 풀고 풀어서 오차를 최소화하는 것
* 다시말해, 기계가 데이터를 기반으로 스스로 학습함으로써 성능이 향상되는 기술이 머신러닝입니다.

# 전통적인 프로그래밍 VS 머신러닝의 차이점

* 프로그래밍 방식: 공식을 알려주고 답을 구하게 한 것
    * 규칙을 세우고 이를 프로그램으로 구현한 후, 데이터를 입력하여 해답을 찾음
    * 변화가 생기면 다시 규칙을 세워야 해서 변화 대응이 어려움
* 기계학습 방식: 답을 알려주고 공식을 깨우치게 한 것
    * 어렸을 때 한글 익히는 방식과 동일 
    * 강아지 사진 또는 실물은 데이터(data)이며, '강아지'라는 이름은 레이블(label)임.

# 기계학습 방법

* 머신러닝 방법은 알고리즘과 데이터를 입력하는 형태에 따라 구분합니다.
* 레이블(정답)의 유무에 따라
* 레이블이 있으면 **지도 학습**
* 레이블이 없으면,**비지도 학습**
* 행동에 따라 보상하여 인공지능을 향상시키는 **강화학습**이 있습니다.


* **```지도 학습```** : 문제와 답을 가르쳐 주면서 공부시키는 방식(친절함)
* **```비지도 학습```**:문제만 잔뜩 주고 알아서 답을 찾으라고 하는 방식(불 친절함)
* **```강화 학습```**: 문제도 답도 아무것도 안 주고, 그냥 컴퓨터를 어떤 장소(환경)에 집어 넣고, 그러면 컴퓨터는 자기 마음대로 행동하게 되는데, 그렇게 행동을 할 때마다 상점 또는 벌점을 줘서 채점을 합니다. 그러면 컴퓨터는 점점 상점을 많이 받는 방식으로 스스로 진화하는 방식
