## 기술 통계(Description Statistics)

* 기술은 Technology가 아닌 Description(묘사)의 의미입니다. 
* 통계적인 방법을 활용해 수집한 데이터를 요약하여 묘사하고 설명하는 기법을 말합니다. 

```
평균과 중간값
- 평균은 이상치 특잇값(outlier)에 큰 영향을 받음
- EDA 에서는 특잇값에 큰 영향을 받지 않는 중간값을 선호
```
